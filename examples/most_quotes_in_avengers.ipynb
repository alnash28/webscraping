## How to figure out which character has the most lines in the avengers based on transcript data

We'll be looking at the transcript from this website [https://www.imsdb.com/scripts/Avengers,-The-(2012).html](https://www.imsdb.com/scripts/Avengers,-The-(2012).html)

```python
import requests
from bs4 import BeautifulSoup
import operator
```

```python
# create the soup object 
url = 'https://www.imsdb.com/scripts/Avengers,-The-(2012).html'
r  = requests.get(url)
data = r.text
soup = BeautifulSoup(data)
```

```python




















































lines_dictionary = {}

for tag in soup.find_all('b'):
    character = tag.get_text().strip()
    if character not in lines_dictionary:
        lines_dictionary[character] = 1
    else:
        lines_dictionary[character] += 1
        
lines_dictionary = dict( sorted(lines_dictionary.items(), key=operator.itemgetter(1),reverse=True))

for character in lines_dictionary:
    print(character,':',lines_dictionary[character])
```
